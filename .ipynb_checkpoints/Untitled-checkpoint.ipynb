```python
data_ar = np.array(data)
m, n = data_ar.shape
np.random.shuffle(data_ar)
print(f"shape of the array:{m}*{n}")

data_thou = data_ar[0:1000].T
y_thou = data_thou[0]
x_thou = data_thou[1:n]

data_train = data_ar[1000:m].T
y_train = data_train[0] / 255
x_train = data_train[1:n] /255
```

```python
## Making a nueral network

def init_pramas():
    W1 = np.random.rand(10, 784)-0.5
    b1 = np.random.rand(10, 1)-0.5
    W2 = np.random.rand(10, 10)-0.5
    b2 = np.random.rand(10, 1)-0.5
    return W1, b1, W2, b2,

def ReLu(z):
    return np.maximum(z, 0)

def softmax(z):
    exp = np.exp(z- np.max(z))# this will give us the probablity
    exp = exp / exp.sum(axis=0)
    return exp

def forward_prop(W1, b1, W2, b2, x):
    Z1 = W1.dot(x) + b1 # this makes up the connections to the hidden layer
    A1 = ReLu(Z1) 
    Z2 = W2.dot(A1) + b2
    A2 = softmax(Z2) # now this will give us the output layer of the neural network
    return Z1, A1, Z2, A2

def one_hot(y):
    one_hot_y = np.zeros((y.size, y.max()+1))
    one_hot_y[np.arange(y.size), y] = 1 
    one_hot_y = one_hot_y.T
    return one_hot_y

def derv_ReLu(z):
    return z> 0

def back_prop(Z1, A1, Z2, A2, W2, x, y):
    m = y.size
    one_hot_Y = one_hot(y)
    dZ2 = A2 - one_hot_Y
    dW2 = 1/ m * dZ2.dot(A1.T)
    db2 = 1/m * np.sum(dZ2)
    dZ1 = W2.T.dot(Z2) * derv_ReLu(Z1)
    dW1 = 1/m * dZ1.dot(x.T)
    db1 = 1/m * np.sum(dZ1)
    return dW1, db1, dW2, db2

def update_prams(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    W1 = W1 - alpha * dW1
    b1 = b1 - alpha * db1
    W2 = W2 - alpha * dW2
    b2 = b2 - alpha * db2
    return W1, b1, W2, b2

```

```python
def get_pred(A2):
    return np.argmax(A2, 0)

def get_acc(pred, y):
    print(pred, y)
    return np.sum(pred == y) / y.size

def gradient_des(x, y, iterations, alpha):
    W1, b1, W2, b2 = init_pramas()
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, x)
        dW1, db1, dW2, db2 = back_prop(Z1, A1, Z2, A2, W2, x, y)
        W1, b1, W2, b2 = update_prams(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        if (i % 10 == 0):
            print("iteration:", i)
            print("accuracy:", get_acc(get_pred(A2), y))
    return W1, b1, W2, b2
```

```python
W1, b1, W2, b1 = gradient_des(x_train, y_train, 500, 0.10)
```
